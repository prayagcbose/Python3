```python
import tkinter as tk
import random
import time

window = tk.Tk()
window.geometry("800x600")

lbl = tk.Label(window,text="Make a choice",fg="black",bg="red",width=30,font=("courier",40))
lbl.pack()

lbl_1 = tk.Label(window,text="",fg="white",bg="black",width=100,font=("courier",40))
lbl_1.pack()

lbl_2 = tk.Label(window,text="",fg="white",bg="red",width=100,
                 font=("courier",40))
lbl_2.pack()

#function to handle choice

def handleChoice(user_choice):
    computer_choice = random.choice(["Rock","Paper","Scissor"])
    lbl_1.configure(text=f"opponent: {computer_choice}")
    lbl_2.configure(text=f"You: {user_choice}")
    result_compair_dict = {
                           "Rock":"Scissor",
                           "Paper":"Rock",
                           "Scissor":"Paper"
                          }
    if result_compair_dict[user_choice] == computer_choice:
        lbl.configure(text="you win :)",
                      fg="green",
                      font=("courier",60,"bold"))
    elif result_compair_dict[computer_choice] == user_choice:
        lbl.configure(text="you loose :(",
                      fg="black",
                      font=("courier",60,"bold"))
    else:
        lbl.configure(text="you tied :|",
                      fg="white",
                      font=("courier",60,"bold"))


rock = tk.Button(window,
                 text="Rock",
                 fg="white",
                 bg="Black",
                 width=10,
                 font=("courier",40),
                 command=lambda:handleChoice("Rock"))
rock.pack()

paper = tk.Button(window,
                  text="Paper",
                  fg="black",
                  bg="red",
                  width=10,
                  font=("courier",40),
                  command=lambda:handleChoice("Paper"))
paper.pack()

scissor = tk.Button(window,
                    text="Scissor",
                    fg="white",
                    bg="Black",
                    width=10,
                    font=("courier",40),
                    command=lambda:handleChoice("Scissor"))
scissor.pack()
window.mainloop()
```
